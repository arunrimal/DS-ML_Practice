```python
movies = {
    'romantic':207,
    'drama':200,
    'action':180,
    'comedy':205
}
```

```python
plt.bar(movies.keys(),movies.values())
```

```python
x_val = np.arange(len(movies.keys()))

x_val
```

```python
plt.yticks(x_val, movies.keys() )
```

```python
for index, val in enumerate(x_val):
    plt.barh(val, list(movies.values())[index], label = list(movies.keys())[index])
plt.yticks(x_val, movies.keys())

y=0
for index, value in movies.items():
    plt.text(value+1,y, value)
    y+=1

plt.title("Freq of Movies", fontsize=20)
plt.legend()
plt.show()    
```

```python
plt.barh(np.arange(len(movies.keys())), movies.values())
plt.yticks(np.arange(len(movies.keys())), movies.keys())
```
